```python
# CELL3 – 5 marks

import numpy as np

print('mean of Academic Reputation = ', round(mean(unis['Academic Reputation']), 2))
print('minimum of Academic Reputation = ', min(unis['Academic Reputation']))
print('maximum of Academic Reputation = ', max(unis['Academic Reputation']))
print('standard deviation of Academic Reputation = ', round(stats.tstd(unis['Academic Reputation']),2))

err = 1.96 * stats.tstd(unis['Academic Reputation']) / np.sqrt(len(unis['Academic Reputation']))

print('95% confidence interval of Academic Reputation = ', round(err, 2))
```

```python
# CELL4 – 9 marks

import matplotlib.pyplot as plt

unis = pd.read_excel('./2018-QS-World-University-Rankings-Top200.xlsx', index_col='Rank')
del unis.index.name

# x = unis['Location'].unique()
x = ['United States', 'United Kingdom', 'Canada', 'Australia', 'Netherlands', 'Germany']
y = []

for location in x:
    tmp = unis.loc[unis['Location'] == location]
    y.append(mean(tmp['Citations per Faculty']))

plt.bar(x, y)

plt.xlabel("Location")
plt.ylabel("Citations per Faculty")
plt.title("Citations per Faculty vs Location")

plt.show()
```
